# Quoter Test — Phase 1 Market Making
Quick sanity check for the `Quoter` class: prices, spreads, inventory and hedge logic.

```python
import sys, os
sys.path.append(os.path.abspath('../src'))

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from utils.stock_simulation import Stock
from utils.market_simulator import Market
from utils.market_maker.quoter import Quoter, QuoterConfig
```

## 1. Setup — Stock + Markets B and C

```python
stock = Stock(drift=0.0, vol=0.10, origin=1.10, tick_size=0.0001)
stock.simulate_gbm(n_days=1, dt_seconds=0.05)

market_B = Market(stock)
market_B.generate_noised_mid_price(vol_factor=0.05)
market_B.build_spread()

market_C = Market(stock)
market_C.generate_noised_mid_price(vol_factor=0.05)
market_C.build_spread()

cfg = QuoterConfig(gamma=0.1, k=1.5, Q_base=100_000.0)
quoter = Quoter(market_B, market_C, config=cfg, capital_K=1_000_000.0)

print(f'Simulation: {stock.n_steps:,} steps | dt={stock.time_step*1000:.0f}ms | lag_B={quoter._lag_B} steps | lag_C={quoter._lag_C} steps')
print(f'Effective latency gap: {quoter._effective_gap_s*1000:.1f}ms')
```

## 2. Run the quoter over the full day

```python
# Sample every 100 steps to keep it fast
sample_every = 100
records = []

for step in range(0, stock.n_steps, sample_every):
    t = stock._time_grid[step]
    quotes, cancel_ids = quoter.compute_quotes(step, t)
    fake_ids = list(range(step * 20, step * 20 + len(quotes)))
    quoter.update_live_ids(fake_ids)
    snap = quoter.snapshot(step, t)
    snap['n_quotes'] = len(quotes)
    records.append(snap)

df = pd.DataFrame(records)
df['t_hours'] = df['t_seconds'] / 3600
print(f'Collected {len(df)} snapshots')
df[['t_hours','fair_mid','reservation_price','best_bid','best_ask','spread_bps','spread_AS','spread_latency']].head(10)
```

## 3. Price levels over time

```python
fig, ax = plt.subplots(figsize=(14, 5))
fig.patch.set_facecolor('#111111')
ax.set_facecolor('#111111')

ax.plot(df['t_hours'], df['fair_mid'],           color='#ffffff', linewidth=1.0, label='Fair mid')
ax.plot(df['t_hours'], df['reservation_price'],  color='#afa9ec', linewidth=1.0, label='Reservation price', linestyle='--')
ax.plot(df['t_hours'], df['best_bid'],            color='#5dcaa5', linewidth=0.8, label='Best bid (A)')
ax.plot(df['t_hours'], df['best_ask'],            color='#ef9f27', linewidth=0.8, label='Best ask (A)')
ax.fill_between(df['t_hours'], df['best_bid'], df['best_ask'], alpha=0.15, color='#ffffff')

ax.set_title('Fair mid, Reservation price and Quotes on Exchange A', color='white', fontsize=13)
ax.set_xlabel('Time (hours)', color='white')
ax.set_ylabel('EUR/USD', color='white')
ax.tick_params(colors='white')
ax.legend(facecolor='#222222', edgecolor='#444444', labelcolor='white')
ax.grid(True, linestyle='--', linewidth=0.4, alpha=0.5, color='#444444')
for spine in ax.spines.values(): spine.set_color('#444444')
plt.tight_layout()
plt.show()
```

## 4. Spread decomposition over time

```python
fig, ax = plt.subplots(figsize=(14, 4))
fig.patch.set_facecolor('#111111')
ax.set_facecolor('#111111')

spread_AS_bps      = df['spread_AS']      * 10000
spread_latency_bps = df['spread_latency'] * 10000

ax.stackplot(df['t_hours'],
             spread_AS_bps,
             spread_latency_bps,
             labels=['AS spread (γσ²(T-t) + liquidity term)', 'Latency premium (2σ√gap)'],
             colors=['#ef9f27', '#f09595'],
             alpha=0.8)

ax.set_title('Spread decomposition over the trading day (bps)', color='white', fontsize=13)
ax.set_xlabel('Time (hours)', color='white')
ax.set_ylabel('Spread (bps)', color='white')
ax.tick_params(colors='white')
ax.legend(facecolor='#222222', edgecolor='#444444', labelcolor='white')
ax.grid(True, linestyle='--', linewidth=0.4, alpha=0.5, color='#444444')
for spine in ax.spines.values(): spine.set_color('#444444')
plt.tight_layout()
plt.show()
```

## 5. Quote levels snapshot — one step

```python
# Reset quoter and get full 20 quotes at midday
quoter2 = Quoter(market_B, market_C, config=cfg, capital_K=1_000_000.0)
midday_step = stock.n_steps // 2
t_mid = stock._time_grid[midday_step]
quotes, _ = quoter2.compute_quotes(midday_step, t_mid)

bids = [(q.price, q.size) for q in quotes if q.direction == 'buy']
asks = [(q.price, q.size) for q in quotes if q.direction == 'sell']

bid_prices, bid_sizes = zip(*bids)
ask_prices, ask_sizes = zip(*asks)

fig, ax = plt.subplots(figsize=(12, 5))
fig.patch.set_facecolor('#111111')
ax.set_facecolor('#111111')

ax.barh(range(10), bid_sizes, color='#5dcaa5', alpha=0.8, label='Bid sizes')
ax.barh(range(10), [-s for s in ask_sizes], color='#ef9f27', alpha=0.8, label='Ask sizes')

bid_labels = [f'{p:.6f}' for p in bid_prices]
ask_labels = [f'{p:.6f}' for p in ask_prices]
ax.set_yticks(range(10))
ax.set_yticklabels([f'L{i+1}: bid={b} / ask={a}' for i,(b,a) in enumerate(zip(bid_labels, ask_labels))], color='white', fontsize=8)

ax.axvline(0, color='white', linewidth=0.8)
ax.set_title(f'10-level quotes on Exchange A at t={t_mid/3600:.2f}h', color='white', fontsize=13)
ax.set_xlabel('Size (EUR)  ← Ask | Bid →', color='white')
ax.tick_params(colors='white')
ax.legend(facecolor='#222222', edgecolor='#444444', labelcolor='white')
ax.grid(True, linestyle='--', linewidth=0.4, alpha=0.4, color='#444444')
for spine in ax.spines.values(): spine.set_color('#444444')
plt.tight_layout()
plt.show()
```

## 6. Inventory and hedge trigger test

```python
quoter3 = Quoter(market_B, market_C, config=cfg, capital_K=1_000_000.0)

inventories = np.linspace(0, 1_000_000, 50)
hedge_flags = [abs(q) / 1_000_000 > cfg.delta_limit for q in inventories]

fig, ax = plt.subplots(figsize=(12, 4))
fig.patch.set_facecolor('#111111')
ax.set_facecolor('#111111')

ax.plot(inventories / 1e6, [abs(q) / 1_000_000 for q in inventories], color='#85b7eb', linewidth=2, label='|q| / K')
ax.axhline(cfg.delta_limit, color='#f09595', linewidth=1.5, linestyle='--', label=f'Delta limit ({cfg.delta_limit:.0%})')
ax.fill_between(inventories / 1e6, cfg.delta_limit, [abs(q) / 1_000_000 for q in inventories],
                where=hedge_flags, color='#f09595', alpha=0.3, label='Hedge zone')

ax.set_title('Delta risk limit — hedge trigger zone', color='white', fontsize=13)
ax.set_xlabel('Inventory (M EUR)', color='white')
ax.set_ylabel('|q| / K', color='white')
ax.tick_params(colors='white')
ax.legend(facecolor='#222222', edgecolor='#444444', labelcolor='white')
ax.grid(True, linestyle='--', linewidth=0.4, alpha=0.5, color='#444444')
for spine in ax.spines.values(): spine.set_color('#444444')
plt.tight_layout()
plt.show()

quoter3.update_inventory(850_000)
print(f'Inventory : {quoter3.inventory:,.0f} EUR')
print(f'|q| / K   : {abs(quoter3.inventory)/1_000_000:.1%}')
print(f'Needs hedge: {quoter3.needs_hedge()}')
print(f'Hedge size : {quoter3.hedge_size():,.0f} EUR')
```
